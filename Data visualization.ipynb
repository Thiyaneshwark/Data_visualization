```python
pivot_table = pd.crosstab(df['home_ownership'], df['loan_status'])

row_sums = pivot_table.sum(axis=1)
percentage_table = pivot_table.div(row_sums, axis=0) * 100

red_shades = ['#89CFF0', '#088F8F', '#5F9EA0']
ax = percentage_table.plot(kind='bar', stacked=True, color=red_shades, figsize=(10, 6))
ax.set_xlabel('Home Ownership')
ax.set_ylabel('Percentage')
ax.set_title('Stacked Bar Chart: Home Ownership vs Loan Status', pad=20)  # Increase the pad value
plt.xticks(rotation=0)
plt.legend(title='Loan Status')
for container in ax.containers:
    ax.bar_label(container, labels=[f"{val:.1f}%" for val in container.datavalues], label_type='edge', color='black', fontsize=10)
plt.tight_layout()
plt.show()
```

```python

# Assuming 'df' is your DataFrame containing the loan data

# Create a pivot table
pivot_table = pd.crosstab(df['term'], df['loan_status'])

# Calculate row sums
row_sums = pivot_table.sum(axis=1)

# Calculate the percentage table
percentage_table = pivot_table.div(row_sums, axis=0) * 100

# Define color shades
blue_shades = ['#89CFF0', '#088F8F', '#5F9EA0']

# Create the stacked bar chart
ax = percentage_table.plot(kind='bar', stacked=True, color=blue_shades, figsize=(10, 6))

# Set labels and title
ax.set_xlabel('Loan Duration')
ax.set_ylabel('Percentage')
ax.set_title('Stacked Bar Chart: Loan Duration vs Loan Status', pad=20)
plt.xticks(rotation=0)

# Add legend
plt.legend(title='Loan Status')

# Add data labels
for container in ax.containers:
    ax.bar_label(container, labels=[f"{val:.1f}%" for val in container.datavalues], label_type='edge', color='black', fontsize=10)

# Adjust layout and display the chart
plt.tight_layout()
plt.show()
```

10+years=management lvl 
1-4years= entry lvl 
5-9 =mid lvl

```python
def map_emp_length_category(emp_length):
    if emp_length == "10+ years":
        return "Management Level"
    elif emp_length in ["1 year", "2 years", "3 years", "4 years"]:
        return "Entry Level"
    elif emp_length in ["5 years", "6 years", "7 years", "8 years", "9 years"]:
        return "Mid Level"
    else:
        return "Unknown"

# Apply the function to create a new column
df['emp_category'] = df['emp_length'].apply(map_emp_length_category)
```

```python
df["emp_category"].value_counts()
```
